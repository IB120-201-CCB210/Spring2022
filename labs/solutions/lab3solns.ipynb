```python
import math as m
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
x = sym.symbols("x",cls=sym.Function)
a, t = sym.symbols("a t")
eq = sym.Eq(sym.diff(x(t),t),a*x(t)**2)
eq
```

```python
sym.dsolve(eq, x(t), ics={x(0):1})
```

```python
#part 3

def LV(t, y):
    N, P = y
    dNdt = a*N - b*N*P #prey
    dPdt = c*N*P - d*P #predator
    return [dNdt, dPdt]

```

```python
#part 4

a = 1.8 
b = 0.4 
c = 0.48
d = 1

initial_condition = [10, 10]
tspan = [0, 100]
soln = solve_ivp(
    LV, 
    tspan, 
    initial_condition) 
    
plt.figure(figsize=(8,5), dpi=100)
plt.plot(soln.t, soln.y[0],label="N")
plt.plot(soln.t, soln.y[1],label="P")
plt.xlabel("Time, $t$", fontsize=20)
plt.ylabel("Population Size", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.legend(fontsize=15);
```

```python
#part 5

N, P, a, b, c, d = sym.symbols("N, P, a, b, c, d")
dNdt = a*N - b*N*P #prey
dPdt = c*N*P - d*P

sym.nonlinsolve(
    [dNdt, dPdt],
    [N,P]
)
```
